## **Creating Dataset**

```python
import numpy.random as npr
data = np.exp ( npr.randn(50, 5))
data
```

# **Mean and Standard Deviation**

```python
dmean=data.mean(0)
dmean
```

```python
dstd=data.std(axis=0)
dstd
```

# Standardizing Data

#Using given formula Standardizing data

```python
normalized = (data - dmean)/(dstd)
normalized
```

```python
normalized.mean(0)
```

```python
normalized.std(0)
```

**Standardizing data by importing libraries**

```python
from sklearn import preprocessing
import numpy as np
scaler = preprocessing.StandardScaler().fit(data)
scaler
X_scaled = scaler.transform(data)
X_scaled
```

```python
X_scaled.mean(axis=0)
```

```python
X_scaled.std(axis=0)
```
